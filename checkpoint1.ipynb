```python
def mergeSort(arr):
    if len(arr) > 1:
 
         # Finding the mid of the array
        mid = len(arr)//2
 
        # Dividing the array elements
        L = arr[:mid]
 
        # into 2 halves
        R = arr[mid:]
 
        # Sorting the first half
        mergeSort(L)
 
        # Sorting the second half
        mergeSort(R)
        print(R,L)
        print("okok")
        i = j = k = 0
 
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            print(arr)
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
 
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
```

```python
# function for removing common characters
# with their respective occurrences
def remove_match_char(list1, list2): 
    for i in range(len(list1)) :
        for j in range(len(list2)) :
 
            # if common character is found
            # then remove that character
            # and return list of concatenated
            # list with True Flag
            if list1[i] == list2[j] :
                print(list1[i])
                c = list1[i]
 
                # remove character from the list
                list1.remove(c)
                list2.remove(c)

                # concatenation of two list elements with *
                # * is act as border mark here
                list3 = list1 + ["*"] + list2
 
                # return the concatenated list with True flag
                return [list3, True]
 
    # no common characters is found
    # return the concatenated list with False flag
    list3 = list1 + ["*"] + list2
    return [list3, False]
```

```python
nom1 = "AJAY"
nom2 = "PRIYA"
remove_match_char(list(nom1),list(nom2))
```

```python
def removeChar(char1,char2):
    i = 0
    j = 0
    while(i<len(char1)):
        if char1[i] in char2:
            char2.remove(char1[i])
            char1.remove(char1[i])
        else:
            i+=1
    while(j<len(char2)):
        if char2[i] in char1:
            char2.remove(char2[j])
            char1.remove(char2[j])
        else:
            j+=1
    return len(char1)+len(char2)

removeChar(list(nom1),list(nom2))           
```

```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
def removeChar2(char1,char2):
    inter = list(set(intersection(char1,char2)))
    for ch in inter:
        char1.remove(ch)
        char2.remove(ch)
    return len(char1)+len(char2)
removeChar2(list(nom1),list(nom2))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import svm, datasets
from sklearn.metrics import accuracy_score


X = df[["Pclass","Age","SibSp","Fare"]].values
y = df["Survived"].values
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size=0.2, random_state=42)#stratify
parameters = {'kernel':("poly",'linear', 'rbf'), 'C':[1, 10],'degree':[2,5]}
svc = svm.SVC() # SVM algorithm for classification
clf = GridSearchCV(estimator = svc, param_grid = parameters,scoring="accuracy") # performance mesures= accuracy/recall/precision/F1-score //// AUCROC
grid_result = clf.fit(X_train, y_train.reshape(-1,1))
grid_result.best_score_,grid_result.best_params_


#print('r_sqr value:',reg.score(X_test, y_test))
```
